```python
# ============================================================
# ENTERPRISE FINANCE KNOWLEDGE GRAPH EXPLORER
# SAP FICO-STYLE ENTITY RELATIONSHIPS + IMPACT SEARCH
# RUN THIS ENTIRE CELL IN GOOGLE COLAB
# ============================================================

# =========================
# 1. INSTALL DEPENDENCIES
# =========================
!pip -q install pandas numpy networkx matplotlib scikit-learn

# =========================
# 2. IMPORT LIBRARIES
# =========================
import warnings
warnings.filterwarnings("ignore")

import random
from datetime import datetime, timedelta
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

random.seed(42)
np.random.seed(42)

pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 50)

PROJECT_DIR = Path("enterprise_finance_knowledge_graph_output")
PROJECT_DIR.mkdir(exist_ok=True)

print("Project directory:", PROJECT_DIR.resolve())

# =========================
# 3. MASTER DATA CREATION
# =========================
def make_vendors(n=90):
    vendor_bases = [
        "Alpha Grid Services", "Northwind Electric Supply", "GreenVolt Systems", "Prime Utility Parts",
        "TransEnergy Logistics", "Vertex Maintenance", "BluePeak Consulting", "SolarEdge Components",
        "Metro Industrial Tools", "Crescent Field Ops", "Pioneer Cable Works", "Helios Analytics",
        "Summit Procurement", "IronBridge Equipment", "EnerSys Contractors", "Utility Edge Networks"
    ]
    categories = ["Maintenance", "IT Services", "Electrical Components", "Logistics", "Consulting", "Field Services"]
    states = ["NY", "CT", "ME", "MA", "PA", "NJ", "RI"]

    rows = []
    for i in range(1, n + 1):
        rows.append({
            "vendor_id": f"V{i:04d}",
            "vendor_name": random.choice(vendor_bases) + f" {i}",
            "vendor_category": random.choice(categories),
            "vendor_state": random.choice(states),
            "risk_rating": random.choice(["Low", "Medium", "High"]),
            "preferred_vendor_flag": random.choice([1, 1, 1, 0]),
            "contract_type": random.choice(["Annual", "Project", "Emergency", "Managed Service"])
        })
    return pd.DataFrame(rows)

def make_cost_centers():
    rows = [
        ("CC100", "Transmission Operations", "Operations"),
        ("CC110", "Distribution Maintenance", "Operations"),
        ("CC120", "Substation Engineering", "Engineering"),
        ("CC130", "Renewables Program", "Strategy"),
        ("CC140", "Customer Support", "Customer Ops"),
        ("CC150", "IT Shared Services", "Technology"),
        ("CC160", "Finance Operations", "Finance"),
        ("CC170", "Vegetation Management", "Field Ops"),
        ("CC180", "Grid Modernization", "Engineering"),
        ("CC190", "Regulatory Affairs", "Corporate")
    ]
    return pd.DataFrame(rows, columns=["cost_center_id", "cost_center_name", "department"])

def make_profit_centers():
    rows = [
        ("PC200", "Electric Distribution East", "East"),
        ("PC210", "Electric Distribution West", "West"),
        ("PC220", "Gas and Field Services", "North"),
        ("PC230", "Renewables and Storage", "Enterprise"),
        ("PC240", "Corporate Shared Services", "Corporate")
    ]
    return pd.DataFrame(rows, columns=["profit_center_id", "profit_center_name", "region_group"])

def make_gl_accounts():
    rows = [
        ("500100", "Opex - Maintenance", "Expense"),
        ("500200", "Opex - IT Services", "Expense"),
        ("500300", "Opex - Contractors", "Expense"),
        ("500400", "Opex - Logistics", "Expense"),
        ("500500", "Opex - Vegetation Mgmt", "Expense"),
        ("500600", "Opex - Consulting", "Expense"),
        ("500700", "Opex - Emergency Repairs", "Expense"),
        ("500800", "Capex - Grid Modernization", "Capital"),
        ("500900", "Capex - Substation Upgrades", "Capital"),
        ("110100", "Accounts Payable", "Liability")
    ]
    return pd.DataFrame(rows, columns=["gl_account", "gl_account_name", "gl_type"])

vendors = make_vendors(90)
cost_centers = make_cost_centers()
profit_centers = make_profit_centers()
gl_accounts = make_gl_accounts()

print("Master data created:")
print("vendors:", vendors.shape)
print("cost_centers:", cost_centers.shape)
print("profit_centers:", profit_centers.shape)
print("gl_accounts:", gl_accounts.shape)

# =========================
# 4. SYNTHETIC TRANSACTIONS
# =========================
descriptions = [
    "substation maintenance support",
    "vegetation clearing around power lines",
    "grid inspection contractor services",
    "cybersecurity monitoring support",
    "cloud cost optimization consulting",
    "field equipment transportation",
    "metering systems enhancement",
    "SCADA support services",
    "storm response contractor mobilization",
    "distribution pole inspection",
    "asset reliability assessment",
    "emergency transformer replacement",
    "temporary field crew support",
    "circuit restoration activities",
    "renewables integration support",
    "substation modernization project"
]

def generate_transactions(num_rows=5500):
    start_date = datetime(2024, 1, 1)
    end_date = datetime(2025, 12, 31)
    date_range_days = (end_date - start_date).days

    rows = []
    for i in range(1, num_rows + 1):
        vendor = vendors.sample(1).iloc[0]
        cc = cost_centers.sample(1).iloc[0]
        pc = profit_centers.sample(1).iloc[0]
        gl = gl_accounts.sample(1).iloc[0]

        posting_date = start_date + timedelta(days=random.randint(0, date_range_days))
        amount = round(random.uniform(4000, 95000), 2)

        if vendor["vendor_category"] == "Consulting":
            amount *= 1.20
        if cc["cost_center_name"] == "Vegetation Management":
            amount *= 1.15
        if gl["gl_account_name"] in ["Opex - Emergency Repairs", "Capex - Substation Upgrades"]:
            amount *= 1.25
        if pc["profit_center_name"] == "Renewables and Storage":
            amount *= 1.10

        rows.append({
            "document_id": f"DOC{i:06d}",
            "invoice_id": f"INV{i:06d}",
            "posting_date": posting_date.date().isoformat(),
            "vendor_id": vendor["vendor_id"],
            "cost_center_id": cc["cost_center_id"],
            "profit_center_id": pc["profit_center_id"],
            "gl_account": gl["gl_account"],
            "amount_usd": round(amount, 2),
            "currency": "USD",
            "description": random.choice(descriptions),
            "status": random.choice(["POSTED", "APPROVED", "PAID", "PENDING_REVIEW"])
        })

    return pd.DataFrame(rows)

finance_transactions = generate_transactions(5500)

# Inject stronger relationship clusters to make graph meaningful
cluster_idx = np.random.choice(finance_transactions.index, size=300, replace=False)
target_vendors = vendors.sample(8, random_state=42)["vendor_id"].tolist()
target_cost_centers = ["CC170", "CC120", "CC180", "CC150"]
target_gls = ["500500", "500700", "500600", "500800"]

finance_transactions.loc[cluster_idx, "vendor_id"] = np.random.choice(target_vendors, size=len(cluster_idx), replace=True)
finance_transactions.loc[cluster_idx, "cost_center_id"] = np.random.choice(target_cost_centers, size=len(cluster_idx), replace=True)
finance_transactions.loc[cluster_idx, "gl_account"] = np.random.choice(target_gls, size=len(cluster_idx), replace=True)
finance_transactions.loc[cluster_idx, "amount_usd"] = np.random.uniform(45000, 160000, size=len(cluster_idx)).round(2)
finance_transactions.loc[cluster_idx, "description"] = "strategic high-value relationship cluster activity"

finance_transactions.to_csv(PROJECT_DIR / "finance_transactions.csv", index=False)

print("\nTransactions created:", finance_transactions.shape)
display(finance_transactions.head())

# =========================
# 5. DATA MODEL JOIN
# =========================
df = (
    finance_transactions
    .merge(vendors, on="vendor_id", how="left")
    .merge(cost_centers, on="cost_center_id", how="left")
    .merge(profit_centers, on="profit_center_id", how="left")
    .merge(gl_accounts, on="gl_account", how="left")
)

df["posting_date"] = pd.to_datetime(df["posting_date"])
df["month"] = df["posting_date"].dt.to_period("M").astype(str)

print("\nJoined model shape:", df.shape)
display(df.head())

# =========================
# 6. ENTITY RELATIONSHIP TABLES
# =========================
vendor_cost_center_rel = (
    df.groupby(["vendor_id", "vendor_name", "cost_center_id", "cost_center_name"], as_index=False)
      .agg(transaction_count=("document_id", "count"),
           total_amount_usd=("amount_usd", "sum"))
)

vendor_gl_rel = (
    df.groupby(["vendor_id", "vendor_name", "gl_account", "gl_account_name"], as_index=False)
      .agg(transaction_count=("document_id", "count"),
           total_amount_usd=("amount_usd", "sum"))
)

cost_center_gl_rel = (
    df.groupby(["cost_center_id", "cost_center_name", "gl_account", "gl_account_name"], as_index=False)
      .agg(transaction_count=("document_id", "count"),
           total_amount_usd=("amount_usd", "sum"))
)

vendor_profit_center_rel = (
    df.groupby(["vendor_id", "vendor_name", "profit_center_id", "profit_center_name"], as_index=False)
      .agg(transaction_count=("document_id", "count"),
           total_amount_usd=("amount_usd", "sum"))
)

# Save relationship tables
vendor_cost_center_rel.to_csv(PROJECT_DIR / "vendor_cost_center_relationships.csv", index=False)
vendor_gl_rel.to_csv(PROJECT_DIR / "vendor_gl_relationships.csv", index=False)
cost_center_gl_rel.to_csv(PROJECT_DIR / "cost_center_gl_relationships.csv", index=False)
vendor_profit_center_rel.to_csv(PROJECT_DIR / "vendor_profit_center_relationships.csv", index=False)

# =========================
# 7. BUILD KNOWLEDGE GRAPH
# =========================
G = nx.Graph()

# Add nodes
for _, row in vendors.iterrows():
    G.add_node(row["vendor_id"], label=row["vendor_name"], node_type="Vendor", metadata=row.to_dict())

for _, row in cost_centers.iterrows():
    G.add_node(row["cost_center_id"], label=row["cost_center_name"], node_type="Cost Center", metadata=row.to_dict())

for _, row in profit_centers.iterrows():
    G.add_node(row["profit_center_id"], label=row["profit_center_name"], node_type="Profit Center", metadata=row.to_dict())

for _, row in gl_accounts.iterrows():
    G.add_node(row["gl_account"], label=row["gl_account_name"], node_type="GL Account", metadata=row.to_dict())

# Add edges from transactions
for _, row in df.iterrows():
    vendor = row["vendor_id"]
    cc = row["cost_center_id"]
    pc = row["profit_center_id"]
    gl = row["gl_account"]
    amt = float(row["amount_usd"])

    def add_or_update_edge(a, b, relationship_type):
        if G.has_edge(a, b):
            G[a][b]["transaction_count"] += 1
            G[a][b]["total_amount_usd"] += amt
        else:
            G.add_edge(a, b,
                       relationship_type=relationship_type,
                       transaction_count=1,
                       total_amount_usd=amt)

    add_or_update_edge(vendor, cc, "vendor_to_cost_center")
    add_or_update_edge(vendor, gl, "vendor_to_gl")
    add_or_update_edge(cc, gl, "cost_center_to_gl")
    add_or_update_edge(vendor, pc, "vendor_to_profit_center")

print("\nKnowledge graph built.")
print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())

# =========================
# 8. GRAPH ANALYTICS
# =========================
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G, k=min(50, len(G.nodes())), seed=42)

node_summary = []
for node_id, attrs in G.nodes(data=True):
    node_summary.append({
        "node_id": node_id,
        "label": attrs["label"],
        "node_type": attrs["node_type"],
        "degree_centrality": degree_centrality.get(node_id, 0),
        "betweenness_centrality": betweenness_centrality.get(node_id, 0),
        "degree": G.degree(node_id)
    })

node_summary_df = pd.DataFrame(node_summary).sort_values(
    ["betweenness_centrality", "degree_centrality"], ascending=False
)

top_influential_nodes = node_summary_df.head(25).copy()
top_influential_nodes.to_csv(PROJECT_DIR / "top_influential_nodes.csv", index=False)

print("\nTop influential nodes")
display(top_influential_nodes)

# =========================
# 9. IMPACT ANALYSIS FUNCTIONS
# =========================
def get_node_label_map(graph):
    return {node: data["label"] for node, data in graph.nodes(data=True)}

node_label_map = get_node_label_map(G)

def get_connected_entities(node_id, max_results=20):
    if node_id not in G.nodes:
        return pd.DataFrame(columns=["source_node", "connected_node", "connected_label", "connected_type", "relationship_type", "transaction_count", "total_amount_usd"])

    rows = []
    for neighbor in G.neighbors(node_id):
        edge = G[node_id][neighbor]
        rows.append({
            "source_node": node_id,
            "source_label": G.nodes[node_id]["label"],
            "connected_node": neighbor,
            "connected_label": G.nodes[neighbor]["label"],
            "connected_type": G.nodes[neighbor]["node_type"],
            "relationship_type": edge["relationship_type"],
            "transaction_count": edge["transaction_count"],
            "total_amount_usd": round(edge["total_amount_usd"], 2)
        })

    result = pd.DataFrame(rows).sort_values(
        ["total_amount_usd", "transaction_count"], ascending=False
    ).head(max_results)

    return result

def impact_analysis(node_id):
    connections = get_connected_entities(node_id, max_results=100)
    if connections.empty:
        return {
            "node_id": node_id,
            "label": None,
            "node_type": None,
            "connected_entities": 0,
            "total_linked_amount_usd": 0,
            "high_value_connections": 0
        }

    return {
        "node_id": node_id,
        "label": G.nodes[node_id]["label"],
        "node_type": G.nodes[node_id]["node_type"],
        "connected_entities": len(connections),
        "total_linked_amount_usd": round(connections["total_amount_usd"].sum(), 2),
        "high_value_connections": int((connections["total_amount_usd"] > 500000).sum())
    }

# Example impact analysis
sample_node = top_influential_nodes.iloc[0]["node_id"]
sample_connections = get_connected_entities(sample_node, max_results=10)
sample_impact = impact_analysis(sample_node)

print("\nSample impact analysis:")
print(sample_impact)
display(sample_connections)

# =========================
# 10. NATURAL LANGUAGE RELATIONSHIP SEARCH
# =========================
def build_entity_document(row):
    return (
        f"Node {row['node_id']} labeled {row['label']} is of type {row['node_type']}. "
        f"It has degree {row['degree']}. "
        f"Degree centrality is {row['degree_centrality']:.6f}. "
        f"Betweenness centrality is {row['betweenness_centrality']:.6f}. "
        f"This entity is important in enterprise finance relationship analysis."
    )

node_summary_df["entity_document"] = node_summary_df.apply(build_entity_document, axis=1)

relationship_documents = []
for node_id in G.nodes():
    connected_df = get_connected_entities(node_id, max_results=10)
    if not connected_df.empty:
        summary_text = " ".join([
            f"Connected to {r.connected_label} ({r.connected_type}) via {r.relationship_type} "
            f"with amount {r.total_amount_usd:.2f} and transaction count {r.transaction_count}."
            for _, r in connected_df.iterrows()
        ])
    else:
        summary_text = "No connected relationships found."

    relationship_documents.append({
        "node_id": node_id,
        "label": G.nodes[node_id]["label"],
        "node_type": G.nodes[node_id]["node_type"],
        "relationship_document": (
            f"Entity {G.nodes[node_id]['label']} of type {G.nodes[node_id]['node_type']}. "
            f"{summary_text}"
        )
    })

relationship_docs_df = pd.DataFrame(relationship_documents)
relationship_docs_df.to_csv(PROJECT_DIR / "relationship_documents.csv", index=False)

vectorizer = TfidfVectorizer(stop_words="english", ngram_range=(1, 2), max_features=6000)
tfidf_matrix = vectorizer.fit_transform(relationship_docs_df["relationship_document"])

def search_relationships(query, top_k=5):
    query_vec = vectorizer.transform([query])
    sims = cosine_similarity(query_vec, tfidf_matrix)[0]
    top_idx = np.argsort(sims)[::-1][:top_k]
    result = relationship_docs_df.iloc[top_idx].copy()
    result["similarity_score"] = sims[top_idx]
    return result.reset_index(drop=True)

sample_query = "high value vendor relationships in vegetation management and emergency repairs"
sample_search_results = search_relationships(sample_query, top_k=5)

print("\nSample relationship search results")
display(sample_search_results)

# =========================
# 11. COMMUNITY / CLUSTER DETECTION
# =========================
communities = list(nx.community.greedy_modularity_communities(G))
community_rows = []

for idx, community in enumerate(communities, start=1):
    for node in community:
        community_rows.append({
            "community_id": f"COMM{idx:03d}",
            "node_id": node,
            "label": G.nodes[node]["label"],
            "node_type": G.nodes[node]["node_type"]
        })

community_df = pd.DataFrame(community_rows)
community_df.to_csv(PROJECT_DIR / "community_assignments.csv", index=False)

community_summary = (
    community_df.groupby("community_id", as_index=False)
      .agg(node_count=("node_id", "count"))
      .sort_values("node_count", ascending=False)
)
community_summary.to_csv(PROJECT_DIR / "community_summary.csv", index=False)

print("\nCommunity summary")
display(community_summary.head(10))

# =========================
# 12. KEY BUSINESS VIEWS
# =========================
top_vendor_cost_center_links = (
    vendor_cost_center_rel.sort_values(["total_amount_usd", "transaction_count"], ascending=False)
    .head(25)
)
top_vendor_gl_links = (
    vendor_gl_rel.sort_values(["total_amount_usd", "transaction_count"], ascending=False)
    .head(25)
)
top_cost_center_gl_links = (
    cost_center_gl_rel.sort_values(["total_amount_usd", "transaction_count"], ascending=False)
    .head(25)
)

top_vendor_cost_center_links.to_csv(PROJECT_DIR / "top_vendor_cost_center_links.csv", index=False)
top_vendor_gl_links.to_csv(PROJECT_DIR / "top_vendor_gl_links.csv", index=False)
top_cost_center_gl_links.to_csv(PROJECT_DIR / "top_cost_center_gl_links.csv", index=False)

print("\nTop vendor-cost center links")
display(top_vendor_cost_center_links.head(10))

# =========================
# 13. VISUALIZATIONS
# =========================
plt.figure(figsize=(12, 5))
plt.bar(top_influential_nodes["label"].head(10), top_influential_nodes["betweenness_centrality"].head(10))
plt.xticks(rotation=45, ha="right")
plt.title("Top Influential Finance Entities by Betweenness Centrality")
plt.ylabel("Betweenness Centrality")
plt.tight_layout()
plt.show()

monthly_amount = (
    df.groupby("month", as_index=False)
      .agg(total_amount_usd=("amount_usd", "sum"))
      .sort_values("month")
)

plt.figure(figsize=(12, 5))
plt.plot(monthly_amount["month"], monthly_amount["total_amount_usd"], marker="o")
plt.xticks(rotation=45, ha="right")
plt.title("Monthly Enterprise Finance Amount Trend")
plt.ylabel("Total Amount USD")
plt.tight_layout()
plt.show()

# =========================
# 14. SAVE BASE FILES
# =========================
vendors.to_csv(PROJECT_DIR / "vendors.csv", index=False)
cost_centers.to_csv(PROJECT_DIR / "cost_centers.csv", index=False)
profit_centers.to_csv(PROJECT_DIR / "profit_centers.csv", index=False)
gl_accounts.to_csv(PROJECT_DIR / "gl_accounts.csv", index=False)
node_summary_df.to_csv(PROJECT_DIR / "node_summary.csv", index=False)

# =========================
# 15. INTERACTIVE SEARCHES
# =========================
queries = [
    "vendors strongly connected to vegetation management",
    "entities related to emergency repairs and high value finance activity",
    "important cost centers connected to consulting and contractors",
    "profit centers connected to grid modernization and substation upgrades",
    "high impact vendor relationships in field operations"
]

for q in queries:
    print("\n" + "=" * 100)
    print("QUERY:", q)
    display(search_relationships(q, top_k=5))

# =========================
# 16. SHOW GENERATED FILES
# =========================
print("\nGenerated files:")
for p in sorted(PROJECT_DIR.glob("*")):
    print("-", p.name)
```
